```python
import matplotlib.pyplot as plt

# 남양주시의 각 읍별로 반려견 시설별 반려인 수 (가정된 데이터)
neighborhoods = ["와부읍", "진접읍", "화도읍", "진건읍", "오남읍"]  # 읍/면/동 이름
registered_pets_per_facility = {
    "병원": [20, 35, 40, 10, 25],
    "미용": [30, 45, 60, 15, 30],
    "약국": [25, 40, 50, 20, 35],
    "용품": [15, 30, 45, 10, 20],
    "위탁": [10, 20, 25, 5, 15]
}

# 정규화 함수 정의
def normalize_data(data_dict):
    max_values = {}
    # 각 읍별 최대값 구하기
    for facility, numbers in data_dict.items():
        for i, num in enumerate(numbers):
            if neighborhoods[i] in max_values:
                max_values[neighborhoods[i]] = max(max_values[neighborhoods[i]], num)
            else:
                max_values[neighborhoods[i]] = num
    
    # 데이터 정규화
    normalized_data = {facility: [] for facility in data_dict}
    for facility, numbers in data_dict.items():
        for i, num in enumerate(numbers):
            normalized_data[facility].append(num / max_values[neighborhoods[i]])
    
    return normalized_data

# 데이터 정규화
normalized_data = normalize_data(registered_pets_per_facility)

# 정규화된 데이터로 막대 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 7)) 

# x 위치 설정
x = range(len(neighborhoods))

# 각 시설별로 막대 그래프 그리기
facility_count = len(normalized_data)
width = 0.8 / facility_count  # 막대의 폭
colors = ['b', 'g', 'r', 'c', 'm']  # 막대의 색상

for i, (facility, values) in enumerate(normalized_data.items()):
    ax.bar(x, values, width=width, label=facility, color=colors[i], align='center', alpha=0.7)
    x = [xi + width for xi in x]

# x축 레이블을 갱신
ax.set_xticks([xi + width * (facility_count / 2) - (width / 2) for xi in range(len(neighborhoods))])
ax.set_xticklabels(neighborhoods)

# 제목과 축 레이블 추가
ax.set_title('Normalized Dog Owners by Facility in Namyangju-eup')
ax.set_xlabel('Neighborhood')
ax.set_ylabel('Normalized Number of Dog Owners')

# 범례 표시
ax.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 남양주시의 각 읍별로 반려견 시설별 반려인 수 (가정된 데이터)
neighborhoods = ["와부읍", "진접읍", "화도읍", "진건읍", "오남읍"]
facility_labels = ["병원", "미용", "약국", "용품", "위탁"]
registered_pets_per_facility = np.array([
    [20, 30, 25, 15, 10],  # 와부읍
    [35, 45, 40, 30, 20],  # 진접읍
    [40, 60, 50, 45, 25],  # 화도읍
    [10, 15, 20, 10, 5],   # 진건읍
    [25, 30, 35, 20, 15]   # 오남읍
])

# 데이터 정규화
normalized_data = registered_pets_per_facility / registered_pets_per_facility.sum(axis=1)[:, np.newaxis]

# 막대 그래프의 여러 부분들이 차지하는 비율을 누적한 값
cumulative_data = normalized_data.cumsum(axis=1)

# 각 시설별 비율에 따른 색상 지정
colors = ['skyblue', 'orange', 'green', 'red', 'purple']

# 막대 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 6))

# 누적 막대 그래프 생성
for i, (color, label) in enumerate(zip(colors, facility_labels)):
    # 첫 번째 막대(section)는 바닥부터 시작하지만, 그 뒤로는 이전 막대의 누적값부터 시작
    starts = cumulative_data[:, i - 1] if i != 0 else np.zeros(len(normalized_data))

    # 막대의 각 섹션을 그림
    ax.bar(neighborhoods, normalized_data[:, i], bottom=starts, color=color, edgecolor='white', label=label)

# 제목과 축 레이블 추가
ax.set_title('Normalized Dog Owners by Facility in Namyangju-eup')
ax.set_xlabel('Neighborhood')
ax.set_ylabel('Percentage of Dog Owners')

# 범례 표시
ax.legend(title='Facility')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 남양주시의 각 읍별로 반려견 시설별 반려인 수 (가정된 데이터)
neighborhoods = ["와부읍", "진접읍", "화도읍", "진건읍", "오남읍"]
facility_labels = ["병원", "미용", "약국", "용품", "위탁"]
registered_pets_per_facility = np.array([
    [20, 30, 25, 15, 10],  # 와부읍
    [35, 45, 40, 30, 20],  # 진접읍
    [40, 60, 50, 45, 25],  # 화도읍
    [10, 15, 20, 10, 5],   # 진건읍
    [25, 30, 35, 20, 15]   # 오남읍
])

# 데이터 정규화
normalized_data = registered_pets_per_facility / registered_pets_per_facility.sum(axis=1)[:, np.newaxis]

# 막대 그래프의 여러 부분들이 차지하는 비율을 누적한 값
cumulative_data = normalized_data.cumsum(axis=1)

# 각 읍별 총 반려인 수 (가정된 데이터)
total_owners = registered_pets_per_facility.sum(axis=1)

# 각 시설별 비율에 따른 색상 지정
colors = ['skyblue', 'orange', 'green', 'red', 'purple']

# 막대 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 6))

# 누적 막대 그래프 생성
for i, (color, label) in enumerate(zip(colors, facility_labels)):
    # 첫 번째 막대(section)는 바닥부터 시작하지만, 그 뒤로는 이전 막대의 누적값부터 시작
    starts = cumulative_data[:, i - 1] if i != 0 else np.zeros(len(normalized_data))

    # 막대의 각 섹션을 그림
    ax.bar(neighborhoods, normalized_data[:, i], bottom=starts, width=0.5, color=color, edgecolor='white', label=label)

# 총 반려인 수에 대한 꺾은선 그래프의 두 번째 y축
ax2 = ax.twinx()
ax2.plot(neighborhoods, total_owners, color='black', marker='o', linestyle='-', linewidth=2, label='Total Owners')
ax2.set_ylabel('Total Number of Dog Owners')

# 제목과 축 레이블 추가
ax.set_title('Normalized Dog Owners by Facility with Total Owners in Namyangju-eup')
ax.set_xlabel('Neighborhood')
ax.set_ylabel('Percentage of Dog Owners')

# 범례 표시. 범례를 하나로 합쳐서 표시
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```
