```python
boundary = nam_boundary[['ADM_NM', 'geometry']].set_index('ADM_NM')
boundary
```

```python
df2 = df1.loc[df1['행정동'] == '다산동']
df2.loc[13, ['행정동']] = '다산1동'
df2
```

```python
df3 = df1.loc[df1['행정동'] == '다산동']
df3.loc[13, ['행정동']] = '다산2동'
df3
```

```python
df4 = pd.concat([df1, df2, df3])
df4 = df4.set_index('행정동')
df4 = df4.drop(['다산동'])
df4
```

```python
df4.to_csv("x_final_normalization_map_visualization.csv", encoding='utf-8-sig')
```

```python
m1 = folium.Map(location=[37.6, 127.15], tiles='cartodbpositron', zoom_start=11)

folium.Choropleth(
    geo_data=boundary['geometry'].geometry.to_crs(epsg=4326).__geo_interface__,
    data=df4['hospital'],
    key_on='feature.id',
    fill_color='YlGnBu',
    legend_name='hospital',
).add_to(m1)

m1.save('visualization_folium_hospital.html')
```
